```python
import pandas as pd

nav = pd.read_csv("../data/raw/02_nav_history - 02_nav_history.csv")

nav["date"] = pd.to_datetime(
    nav["date"],
    dayfirst=True,
    errors="coerce"
)

nav["nav"] = pd.to_numeric(
    nav["nav"],
    errors="coerce"
)

nav = nav.sort_values(
    ["amfi_code","date"]
)

nav = nav.drop_duplicates()

nav["nav"] = nav.groupby(
    "amfi_code"
)["nav"].ffill()

nav = nav[nav["nav"] > 0]

nav.to_csv(
    "../data/processed/nav_history_clean.csv",
    index=False
)
```

```python
perf = pd.read_csv(
    "../data/raw/07_scheme_performance.csv"
)

numeric_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "benchmark_3yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio",
    "sortino_ratio",
    "std_dev_ann_pct",
    "max_drawdown_pct",
    "aum_crore",
    "expense_ratio_pct"
]

for col in numeric_cols:
    perf[col] = pd.to_numeric(
        perf[col],
        errors="coerce"
    )

expense_anomalies = perf[
    (perf["expense_ratio_pct"] < 0.1)
    |
    (perf["expense_ratio_pct"] > 2.5)
]

print(expense_anomalies.shape)

perf.to_csv(
    "../data/processed/scheme_performance_clean.csv",
    index=False
)
```
