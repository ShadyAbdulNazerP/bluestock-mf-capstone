```python
from pathlib import Path

files = Path("../data/raw").glob("*.csv")

for file in files:
    print("\n" + "="*60)
    print(file.name)

    df = pd.read_csv(file)

    print("Shape:", df.shape)
    print("Missing Values:")
    print(df.isnull().sum().sum())

    print("Duplicates:")
    print(df.duplicated().sum())
```
